```python
# 산점도 그래프 (Scatter Plot)
import pandas as pd
mtcars = pd.read_csv("./data/mtcars.csv")
mtcars
```

```python
# EDA 방식으로 데이터 분석
print("데이터 크기:", mtcars.shape) # 32 행 12 열
```

```python
# 앞쪽 데이터 확인
mtcars.head()
```

```python
# 뒤쪽 데이터 확인
mtcars.tail()
```

```python
# 기초 통계량
mtcars.describe()
```

```python
# mpg (y), vs wt(x)  산점도로 시각화
import matplotlib.pyplot as plt

fig = plt.figure() # 틀
axe = fig.add_subplot(1, 1, 1) # 1행 1열의 1번째 축
axe.scatter(mtcars['wt'], mtcars['mpg'])
axe.set_title('mtcats wt vs mpg')
axe.set_xlabel("Weight")
axe.set_ylabel("Miles per gallon")

plt.show()
```

```python
# 상관 계수 
print("상관 계수 (wt, mpg):\n")
mtcars.corr()
```

```python
# 중량 계수 (wt), mpg 의 상관계수
print("wt mpg 상관계수\n", mtcars[['wt', 'mpg']].corr())
# 중량과 연비 간에는 역상관 관계가 성립하는 것으로 추정
```

```python
# boxplot
boxplot = plt.figure()
boxplot_axe = boxplot.add_subplot(1, 1, 1)
boxplot_axe.boxplot([mtcars['mpg'], mtcars['wt']])
boxplot.suptitle('Boxplot of mtcars')
boxplot_axe.set_ylabel('mpg')
plt.show()
```

```python
# IQR 확인 
quartiles = mtcars['mpg'].quantile([0.25, 0.5, 0.75]) # 1사분위수, 2사분위수(중앙값), 3시분위수
first_quartile = quartiles.iloc[0]
third_quartile = quartiles.iloc[2]

first_quartile, third_quartile
```

```python
iqr = third_quartile - first_quartile
print("mpg의 IQR:", iqr)
```

```python
# 3사분위수 + IQR * 1.5 -> 상단 극단치의 경계
# 1사분위수 - IQR * 1.5 -> 하단 극단치의 경계
border_top = third_quartile + iqr * 1.5
border_bottom = first_quartile - iqr * 1.5
print("상단 극단치 경계:", border_top)
print("하단 극단치 경계:", border_bottom)
```

```python
# 상단 극단치 경계를 넘는 mpg 추출 -> 무슨 제품?
mpg_outliers = mtcars[mtcars['mpg'] > border_top]
print("Outliers:")
mpg_outliers
```
